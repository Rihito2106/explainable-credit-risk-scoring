```python
features = ["Age", "Sex", "Job", "Housing", "Saving accounts", "Checking account", "Credit amount","Duration"]
```

```python
target="Risk"
```

```python
df_model = df[features + [target]].copy()
df_model.head()
```

```python
from sklearn.preprocessing import LabelEncoder
import joblib
```

```python
cat_cols = df_model.select_dtypes(include = "object").columns.drop("Risk")

```

```python
le_dict = {}
```

```python
for col in cat_cols:
    le = LabelEncoder()
    df_model[col] = le.fit_transform(df_model[col])
    le_dict[col] = le
    joblib.dump(le, f"{col}_encoder.pkl")
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = df_model.drop(target, axis = 1)
```

```python
y = df_model[target]
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, stratify = y, random_state = 1)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
from sklearn.metrics import average_precision_score, roc_auc_score
!pip install xgboost
import xgboost as xgb
```

```python
# Second cut: From the training set, make a validation set (60% train, 20% validation, 20% test)
X_tr, X_va, y_tr, y_va = train_test_split(X_train, y_train, test_size=0.25, stratify=y_train, random_state=1)
```

```python
# Calculate class imbalance weight (Good vs Bad risk)
neg = int((y_tr == 0).sum())
pos = int((y_tr == 1).sum())
spw = neg / max(pos, 1)
```

```python
print("Train / Val / Test shapes:", X_tr.shape, X_va.shape, X_test.shape)
print(f"neg: {neg}, pos:{pos}, scale_positive_weight (spw):{round(spw, 1)}")
```

```python
# Convert to DMatrix (XGBoost's native data structure)
feat_names = list(X_tr.columns)
dtr = xgb.DMatrix(X_tr, label=y_tr, feature_names=feat_names)
dva = xgb.DMatrix(X_va, label=y_va, feature_names=feat_names)
dte = xgb.DMatrix(X_test, label=y_test, feature_names=feat_names)
```

```python

params = dict(
    objective="binary:logistic",  
    eval_metric="aucpr",          
    tree_method="hist",           
    max_depth=5,
    eta=0.03,                     
    subsample=0.8,
    colsample_bytree=0.8,
    reg_lambda=1.0,
    scale_pos_weight=spw
)

# Train baseline with early stopping
bst = xgb.train(
    params=params,
    dtrain=dtr,
    num_boost_round=100,         # Force it to build exactly 100 trees
    evals=[(dte, "test")],       # Just watch the test set directly
    verbose_eval=False
)
```

```python
cons = [0] * len(feat_names)
if "Credit amount" in feat_names: cons[feat_names.index("Credit amount")] = -1
if "Duration" in feat_names: cons[feat_names.index("Duration")] = -1
if "Age" in feat_names: cons[feat_names.index("Age")] = 1
if "Saving accounts" in feat_names: cons[feat_names.index("Saving accounts")] = 1
if "Checking account" in feat_names: cons[feat_names.index("Checking account")] = 1

mono_string = "(" + ",".join(map(str, cons)) + ")"

params_cons = params.copy()
params_cons.update({"monotone_constraints": mono_string})

bst_cons = xgb.train(
    params=params_cons,
    dtrain=dtr,
    num_boost_round=100,
    evals=[(dte, "test")],
    verbose_eval=False
)
```

```python
p_base = bst.predict(dte)
p_cons = bst_cons.predict(dte)

print("--- MODEL COMPARISON ---")
print("PR AUC  baseline     :", round(average_precision_score(y_test, p_base), 3))
print("PR AUC  constrained  :", round(average_precision_score(y_test, p_cons), 3))
print("ROC AUC baseline     :", round(roc_auc_score(y_test, p_base), 3))
print("ROC AUC constrained  :", round(roc_auc_score(y_test, p_cons), 3))
```
